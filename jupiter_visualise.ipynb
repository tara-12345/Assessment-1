Part 1 - Exploring the data 

1. Reading and summaries

Import the moons class to uitilise its modulues and functions

```python
from jupiter import Moons
moons_data = Moons()
```

In order to get an inital idea of what the data contains, the function "moons_head()" can be used to print the first few lines of data. This shows there are 9 feilds in this data set. 

```python
moons_data.moons_head()
```

From here, it can be seen that the mass column for the moons containts alot of NaN values. To see which moons do have data for this variable, the method "drop_navalues" can be used, including the column name of which to apply this method. 

```python
moons_data.drop_navalues("Mass (kg)")
```

It can then be seen a limited number of jupiters moons contain mass data, and to get a larger summary of all numerical variables, including the count, "data_summaries" can be called from the moons class.


```python
moons_data.data_summaries()
```

From this table, only magnitude and mass have NaN values, as seen from the count. Other data to note would be the standard deviation, where researchers can see how far, on average, each value varies from the respective mean. It can also be inferred there are 56 records in this data set, as this is the maxiumum count value. 

The method column_averages() displays the averages of each column, although this is included in the data_summaries() return, this method presents this key data in a more readable way. This can allow researchers to quickly find the average for any variable in this data set, which is imperative when drawing conclusions about more specific data. 


```python
moons_data.column_average()
```

Furthermore, selecting data for one moon can be done by calling the moon_info() function, this allows researchers to type in their request for which speciifc moon data should be displayed for. The example shows data for the moon Aitne, but if the cell is run again another moon can be chosen. 

```python
moons_data.moon_info()
```

2. Visualising Data 

The "distributions()" method returns histograms for each column in the data set that has numerical value, thus excluding the "Group" column. Period, distance, and magnitiude all present left skewed data, suggesting these values tend to be higher. Conversely, the radius and the eccentricity both show right skewed data, although to a lesser extent, suggesting these values tend to be lower. 

```python
moons_data.distributions()
```

The correlation_heatmap() method displays correlation values between each numerical variable, the darker the colour, the less correlation between the variables. 

From this heatmap, the high correlation between period and distance can be seen. This can be expected as the oribit of a moon that is further away from Jupiter is likely going to last longer than a moon which is closer. Also, the low correlation between magnitude and radius is also expected, as the radius of a moon would not be thought to impact its brightness. 

```python
moons_data.correlation_heatmap()
```

If researchers were to select and print a specific correlation value, they could do so by typing the two variables into the return of the "correlation_value()" method.  but any numerical variable can be used. 

```python
moons_data.correlation_value()
```

The plots for the correlations between any two variables can be done using the "plot()" method, the names of the variables must be included. Period and distance have been used an an example, but any numberical variables can be tested by replacing the variables when calling the method 


```python
moons_data.plot("Distance (km)", "Period (days)")
```

Part 2 - Estimating Jupiters Mass


1. Plotting the Model

the "train.()" method returns a linear regression plot and its r2 value, and the plot of resiudals on the training set to help visulise the difference between the observed and predicted values, so researchers can consider reliability of the regression model. In this case residuals seem to deviate but generally remain around 0, therefore it is a reliable regression model. "train.()" also returns the predicted mass of jupiter, calulated in the moons class using Keplers third law. 

```python
moons_data.train()
```

The literature value for the mass of jupiter is 1.898 × 10^27 kg, this is to the same order of magnitude as the predicted value. This suggests the linear regression model and usin Keplers third law is relaibale and accurate.
